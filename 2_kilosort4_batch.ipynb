### Initialize paths

```python
root_path = 'D:/data/'
experiment = 'clickbait-motivate'
session_type = 'm'

# Get paths to relevant files
data_paths = get_file_paths(f"{root_path}/{experiment}", extension='npy', keyword='mua', session_type=session_type, print_paths=True, print_n=np.inf)
# Root directory for saving Kilosort results
save_path = f"{root_path}/{experiment}/kilosorted/"
```

```python
#data_paths = filter_paths_by_session_id(data_paths, session_id_threshold=7)
data_paths = data_paths[56:57]
data_paths
```

### Kilosort data

```python
settings = {
    'n_chan_bin': 32,
    'Th_universal': 9.0,
    'Th_learned': 8.0,
    'nblocks': 0,
    'drift_smoothing': [0, 0, 0],
    'dmin': 10.0,
    'dminx': 40.0,
    'nearest_chans': 10
}
```

```python
count = 1

# Run kilosort4 on each mua.npy file
for ii, data_path in enumerate(data_paths):
    print(f'Processing mua.npy {ii+1} of {len(data_paths)}')
    results_path = os.path.join(save_path, get_savedirs(data_path))
    kilosort(data_path, results_path, probe_path='probe-maps/8_tetrode_2region_20um_8shank_tight.json', clean_outliers=False, settings=settings)
    clear_output(wait=True)

print('All done!')
print(f'{ii+1} of {len(data_paths)} mua.npy files processed')

# Edit params.py files to point to local temp_wh.dat
params_paths = get_file_paths(f"{root_path}/{experiment}", extension='py', keyword='params')
for params in params_paths:
    reset_dat_path_in_params(params, new_dat_path='temp_wh.dat')

print('params.py files updated to point to local temp_wh.dat')
```
